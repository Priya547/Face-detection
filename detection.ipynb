**1.import the necessary package**

```python
import os
import json
import codecs
import cv2
import matplotlib.pyplot as plt
import numpy as np
import requests
from tqdm import tqdm
from PIL import Image
from io import BytesIO
```

**2. load the images of human_faces**

```python
image_path = '/content/non grp pupil.jpeg'
```

*  we need to read the image with opencv's imread() function*

```python
img = cv2.imread(image_path)
img.shape
```

**3. convert the image to grayscale**

```python
gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray_image.shape
```

**4. load classifier**
*- load the pre-trained Haar Cascade classifier that is built into opencv*

```python
face_classifier = cv2.CascadeClassifier(
    cv2.data.haarcascades + "haarcascade_frontalface_default.xml"
)
```

**5. Face Detection Function**
*- we can now perform face detection on the grayscale image using the classifier*

```python
face = face_classifier.detectMultiScale(
    gray_image, scaleFactor=1.1, minNeighbors=5, minSize=(40, 40)
)
```

**6. Drawing boundary**

```python
for (x, y, w, h) in face:
    cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 4)
```

**7. Display result**

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
import matplotlib.pyplot as plt



plt.figure(figsize=(20,10))
plt.imshow(img_rgb)
plt.axis('off')
```
